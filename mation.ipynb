```python
#林志成
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
lin_c = pd.read_csv("林志成.csv")
cv = pd.read_csv("邱毅恩.csv")

regressor = LinearRegression()

lin1 = lin_c[["成績"]]  
lin2 = lin_c["速度"]  
regressor.fit(lin1, lin2)

w_0 = regressor.intercept_  # 截距
w_1 = regressor.coef_       # 回歸細數
r2 = regressor.score(lin1, lin2)  # R²分數
print("多元回歸模型")
print("截距:", w_0)
print("回歸係數:", w_1)
print("R²分数:", r2)

# 使用模型进行预测
ch1 = cv[["成績"]]
ch2 = cv["速度"]
predictions = regressor.predict(ch1)
```

```python
plt.scatter(lin1,lin2,color="blue",label="score")
plt.plot(lin1,regressor.predict(lin1),color="red",linewidth = 2,label = "V")
plt.title("pole_vault")
plt.xlabel("pole-value score")
plt.ylabel("30m-v")
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("江育安.csv")

df = df.dropna(subset=["成績","速度"])

x1 = df.loc[:, "成績"].values.reshape(-1, 1)
y1 = df.loc[:, "速度"].values

regressor = LinearRegression()
regressor.fit(x1, y1)

w_0 = regressor.intercept_
w_1 = regressor.coef_
r2 = regressor.score(x1, y1)

print("截距:", w_0)
print("回歸系数:", w_1)
print("R²分数:", r2)
```

```python
plt.scatter(x1,y1,color="blue",label="score")
plt.plot(x1,regressor.predict(x1),color="red",linewidth = 2,label = "V")
plt.xlabel("pole-value score")
plt.ylabel("30m-v")
plt.legend()
plt.show()
```
